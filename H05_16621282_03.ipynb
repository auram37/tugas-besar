```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./downloads/H05.csv")
```

```python
df
```

```python
print("Histogram distribusi nilai matematika",df["nilai_mat"].plot.hist())
plt.show()
```

```python
jumlah = df["fakultas"].value_counts()
print("Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas",jumlah.plot.barh())
plt.show()
```

```python
jumlah = df["fakultas"].value_counts()
print("Diagram pie banyaknya mahasiswa masing-masing fakultas",jumlah.plot.pie())
plt.show()
```

```python
print("Berdasarkan diagram pie dan batang horizontal, fakultas dengan mahasiswa terbanyak adalah FTI. Diagram yang lebih baik adalah diagram batang karena memiliki indikator angka yang spesifik, sedangkan pada diagram pie, besarnya persentase antara jumlah mahasiswa STEI dan FTI terlihat hampir sama dan rancu.")
```

```python
df.plot(kind="scatter", x="nilai_kim", y="nilai_fis")
plt.title("Scatter Plot nilai kimia dengan nilai fisika", fontsize=18)
plt.xlabel("nilai kimia", fontsize=14)
plt.ylabel("nilai fisika", fontsize=14)
plt.show()
```
